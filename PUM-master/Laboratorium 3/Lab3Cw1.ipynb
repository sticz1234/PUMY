```python
Wielomianki = np.polyfit(X_Val,Y_Val,18)
Wielomianki  #wspolczynniki wielomianow
```

```python
def Mein(x):
    power = 0 #potegi
    suma = 0 
    for w in reversed(Wielomianki):
        if(power ==0):
            suma += w
        else:
            res = w*x**power
            suma +=res
        power +=1
    return suma 
```

```python
Mein(4000)
```

```python
def do_listy(X):
    lista = []
    for a in range(X.size):
        wyniczek = Mein(a)
        lista.append(wyniczek)
    return lista
```

```python
import csv
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
from sklearn import linear_model,preprocessing
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures
from sklearn.linear_model import LinearRegression
```

```python
data = pd.read_csv("ElectionData.csv") 
data.head(10)
```

```python
data_teritorio = data.query('territoryName == "Território Nacional"')
data_teritorio = data_teritorio.loc[:, ["TimeElapsed","Party","Mandates","FinalMandates"]]
data_teritorio
```

```python
data_teritorio.Party.unique()
```

```python
PS = data_teritorio.query('Party == "PS"')
PPD_PSD = data_teritorio.query('Party == "PPD/PSD"')
BE = data_teritorio.query('Party == "B.E."')
CDS_PP = data_teritorio.query('Party == "CDS-PP"')
PCP_PEV = data_teritorio.query('Party == "PCP-PEV"')
PAN = data_teritorio.query('Party == "PAN"')
CH = data_teritorio.query('Party == "CH"')
RIR = data_teritorio.query('Party == "R.I.R."')
PCTP_MRPP = data_teritorio.query('Party == "PCTP/MRPP"')
A = data_teritorio.query('Party == "A"')
L = data_teritorio.query('Party == "L"')
IL= data_teritorio.query('Party == "IL"')
JPP = data_teritorio.query('Party == "JPP"')
NC = data_teritorio.query('Party == "NC"')
PDR = data_teritorio.query('Party == "PDR"')
PNR = data_teritorio.query('Party == "PNR"')
PURP = data_teritorio.query('Party == "PURP"')
PPM = data_teritorio.query('Party == "PPM"')
MPT = data_teritorio.query('Party == "MPT"')
PTP = data_teritorio.query('Party == "PTP"')
MAS = data_teritorio.query('Party == "MAS"')
```

```python
lista_partii = [PS,PPD_PSD,BE,CDS_PP,PCP_PEV,PAN,CH,RIR,
                    PCTP_MRPP,A,L,IL,JPP,NC,PDR,PNR,PURP,PPM,MPT,PTP,MAS]
```

```python
def partie_funkcja():
    lista_partii = [PS,PPD_PSD,BE,CDS_PP,PCP_PEV,PAN,CH,RIR,
                    PCTP_MRPP,A,L,IL,JPP,NC,PDR,PNR,PURP,PPM,MPT,PTP,MAS]
    lista_wspolczynnikow = []
    lista_predykcji = []
    lista_X_Val = []
    lista_Y_Val = []
    lista_results = []

    for a in range (0,21):
        print(a)
        time = lista_partii[a].iloc[:,0]
        X_Val = time.as_matrix()
        
        mandates = lista_partii[a].iloc[:, 2]
        Y_Val = mandates.as_matrix()
        
        model = Pipeline([('poly', PolynomialFeatures(degree=18)),
                       ('linear', LinearRegression(fit_intercept=True,normalize= True))])
        model = model.fit(X_Val.reshape(-1, 1), Y_Val.reshape(-1, 1))
        
        
        wspolczynniki = model.named_steps['linear'].coef_ 
        predykcje = model.predict(X_Val.reshape(-1, 1))
        
        lista_wspolczynnikow.append(wspolczynniki)
        lista_Y_Val.append(Y_Val)
        
        
        predykcje_df = pd.DataFrame(predykcje).astype(int)
        lista_predykcji.append(predykcje)
        
        X_Val_pd = pd.DataFrame(X_Val)
        lista_X_Val.append(X_Val_pd)
        
        
        lista_Y_Val.append(Y_Val)
        Y_Val_df = pd.DataFrame(Y_Val)
        
        result = pd.concat([X_Val_pd, predykcje_df,Y_Val_df], axis=1, join='inner')
        lista_results.append(result)
        
        a+=1

        
    return lista_wspolczynnikow,lista_predykcji,lista_X_Val,lista_Y_Val,lista_results
    
    
```

```python
lista_wspolczynnikow,lista_predykcji,lista_X_Val,lista_Y_Val,lista_results = partie_funkcja()
```

```python
lista_results[2]  #time, mandaty predykcja, mandaty rzezczywistosc
```

```python
#najwieksze partie - predykcje mandatow
predykcja_partia_PS = lista_results[0].iloc[:, 1].values.tolist()
predykcja_partia_PPD = lista_results[1].iloc[:, 1].values.tolist()
predykcja_partia_BE = lista_results[2].iloc[:, 1].values.tolist()
predykcja_partia_PS
```

```python
def binary_search(item_list,mid):
    first = item_list[0]
    last  = item_list[-1] 
    szukany = last
    print(item_list[mid])
    if szukany - item_list[mid] <= 30 :
        print('got it')
        return item_list[mid],mid
    else:
        if szukany < item_list[mid]:
                mid = mid - 1
                return binary_search(item_list,mid)
                
        else:
                mid = mid + 1
                return binary_search(item_list,mid)
                
```

```python
mid = int(len(predykcja_partia_PS)/2)
a, b = binary_search(predykcja_partia_PS,mid)
wynik_PS = lista_results[0].iloc[b, :]

mid = int(len(predykcja_partia_PPD)/2)
c,d  = binary_search(predykcja_partia_PPD,mid)
wynik_PPD = lista_results[1].iloc[d, :]

mid = int(len(predykcja_partia_BE)/2)
d,e  = binary_search(predykcja_partia_BE,mid)
wynik_BE = lista_results[2].iloc[e, :]
```
