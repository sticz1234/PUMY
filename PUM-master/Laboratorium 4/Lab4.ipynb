```python
   def mod_outlier(df):                             #liczenie IQR
        df1 = df.copy()
        df = df._get_numeric_data()


        q1 = df.quantile(0.25)
        q3 = df.quantile(0.75)
        
        iqr = q3 - q1
        display(iqr)

        lower_bound = q1 -(1.5 * iqr) 
        upper_bound = q3 +(1.5 * iqr)


        for col in df.columns:
            for i in range(0,len(df[col])):
                if df[col][i] < lower_bound[col]:            
                    df[col][i] = 0.0000000000000000001

                if df[col][i] > upper_bound[col]:            
                    df[col][i] = 0.0000000000000000001    


        for col in df.columns:
            df1[col] = df[col]

        return(df1)
```

```python
def Regionowanko(df):
    df = df.copy()
    Regiony = np.unique(abc.values)
    lista_df = []
    for Region in Regiony:
        a = df.loc[df["Location"] == Region]
        lista_df.append(a)
    return lista_df
```

```python
def startyfikacja(lista_region_df):
    lista_modeli = []
    lista_x_test = []
    lista_y_test = []
    lista_accu_region = []
    for region in lista_region_df:
        stratified = StratifiedShuffleSplit(n_splits = 4, test_size=200, random_state = 0)
                                            
        df1 = region.copy()
        Rain = df1['RainTomorrow']
        df1.drop(labels=["RainTomorrow",'Location'],axis=1, inplace=True, errors='raise')
        
        for train_index, test_index in stratified.split(df1,Rain):
            x_train = df1.iloc[train_index]
            y_train = Rain.iloc[train_index]
            x_test =  df1.iloc[test_index]
            y_test = Rain.iloc[test_index]

            log = LogisticRegression(random_state=0,solver='lbfgs')
            log.max_iter = 1000
            log.fit(x_train, y_train)
            
            print(log.n_iter_)
            print(log.score(x_test,y_test))
            
            lista_modeli.append(log)
            lista_x_test.append(x_test)
            lista_y_test.append(y_test)
            lista_accu_region.append(log.score(x_test,y_test))
            
    return lista_modeli,lista_x_test,lista_y_test,lista_accu_region
```

```python
modele,x_test,deszcz_test,lista_accu = startyfikacja(lista_region_df)
```

```python
def sprawdz(modele,x_test,deszcz_test):
    test_size = 200
    XTest = pd.DataFrame()
    YTest = pd.DataFrame()
    lista_wynikow = []
    
    for x in x_test:
        XTest = XTest.append(x.head(test_size))
        
    
    for y in deszcz_test:
        YTest = YTest.append(pd.DataFrame(y).head(test_size))
        
    
    for i in modele:
        lista_wynikow.append(i.score(XTest,YTest))
    return lista_wynikow,XTest,YTest
```

```python
Wyniki,X_test_glob,Y_test_glob = sprawdz(modele,x_test,deszcz_test)     #Wynikowe df z score danych modeli
Wyniki = pd.DataFrame(Wyniki)
Wyniki.sort_values(0)
```
