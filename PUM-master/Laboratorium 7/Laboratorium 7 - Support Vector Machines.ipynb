```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets.samples_generator import make_blobs
from sklearn.model_selection import train_test_split
```

Wylosowanie zbioru pr√≥bek

```python
X, y = make_blobs(n_samples=50, centers=2, n_features=2, random_state=7, cluster_std =2.)
plt.scatter(X[:, 0], X[:, 1], c=y)
plt.xlim(-13,4)
plt.ylim(0,11)
```

Podzia≈Ç pr√≥bek na treningowe i testowe

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.50, random_state=6)

plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
plt.xlim(-13,4)
plt.ylim(0,11)

plt.plot([-9, -2],[ 0, 11])
plt.plot([-5, -5],[ 0, 11])
plt.plot([-7, -1],[11,  0])
```

```python
plt.figure()
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_test)
plt.xlim(-13,4)
plt.ylim(0,11)

plt.plot([-9, -2],[ 0, 11])
plt.plot([-5, -5],[ 0, 11])
plt.plot([-7, -1],[11,  0])
```

Uczenie liniowego klasyfikatora SVM

```python
from sklearn.svm import SVC

model_SVC = SVC(kernel='linear')
model_SVC.fit(X_train, y_train)  

y_pred = model_SVC.predict(X_test)
plt.figure()
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred)
plt.xlim(-13,4)
plt.ylim(0,11)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, y_pred)
```

SVM dla danych nie separowalnych liniowo

```python
from sklearn.datasets import make_moons

X, y = make_moons(noise = 0.1)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.20, random_state=6)
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_train)
```

Uczenie liniowego klasyfikatora SVM

```python
model_SVC = SVC(kernel='linear')
model_SVC.fit(X_train, y_train)  

y_pred = model_SVC.predict(X_test)
plt.figure()
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred)

y_pred_ = model_SVC.predict(X_train)
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_pred_)
```

Uczenie nieliniowego klasyfikatora SVM

```python
model_SVC = SVC(kernel='rbf', gamma='auto')#rbf - radial basis function
model_SVC.fit(X_train, y_train)  

y_pred = model_SVC.predict(X_test)
plt.figure()
plt.scatter(X_test[:, 0], X_test[:, 1], c=y_pred)

y_pred_ = model_SVC.predict(X_train)
plt.figure()
plt.scatter(X_train[:, 0], X_train[:, 1], c=y_pred_)
```

```python
confusion_matrix(y_test, y_pred)
```
